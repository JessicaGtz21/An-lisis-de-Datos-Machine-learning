```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Cargo la imagen mariposaazul
imagen = cv2.imread('/content/mariposaazul.jpg')
```

```python
# Define el kernel para el filtro sharpen
kernel = np.array([[1,2,1],
                   [0,0,0.125],
                   [0.0625, 0.125, 0.625]])
```

```python
# Aplico el filtro sharpen
image_sharpen = cv2.filter2D(imagen, -1, kernel)
```

```python
# Mostrar la imagen original  y la iamgen sharpen
cv2_imshow(imagen)
cv2_imshow(image_sharpen)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
