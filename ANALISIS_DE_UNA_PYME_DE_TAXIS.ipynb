**Distancia entre dos puntos**

```python
import pandas as pd
import folium
from math import radians, sin, cos, sqrt, atan2

# Función para calcular la distancia haversine entre dos puntos
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radio de la Tierra en kilómetros

    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)

    a = sin(dlat / 2) ** 2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distancia = R * c
    return distancia

# Cargar el archivo Excel
url_excel = '/content/CSV_Bases_de_Taxis_Metepec.xlsx'
df = pd.read_excel(url_excel)

# Pedir al usuario que ingrese las coordenadas del primer punto
latitud_punto1 = float(input("Ingrese la latitud del primer punto: "))
longitud_punto1 = float(input("Ingrese la longitud del primer punto: "))

# Pedir al usuario que ingrese las coordenadas del segundo punto
latitud_punto2 = float(input("Ingrese la latitud del segundo punto: "))
longitud_punto2 = float(input("Ingrese la longitud del segundo punto:"))

# Calcular la distancia entre los dos puntos
distancia = haversine(latitud_punto1, longitud_punto1, latitud_punto2, longitud_punto2)

# Crear el mapa centrado en el primer punto ingresado por el usuario
mapa = folium.Map(location=[latitud_punto1, longitud_punto1], zoom_start=12)

# Marcar la ubicación del primer punto en el mapa
folium.Marker([latitud_punto1, longitud_punto1], popup='Primer punto').add_to(mapa)

# Marcar la ubicación del segundo punto en el mapa
folium.Marker([latitud_punto2, longitud_punto2], popup='Segundo punto').add_to(mapa)

# Mostrar la distancia en el mapa
folium.PolyLine([(latitud_punto1, longitud_punto1), (latitud_punto2, longitud_punto2)],
                color='blue', weight=2.5, opacity=1).add_to(mapa)

print(f"Distancia entre los dos puntos: {distancia:.2f} km")

# Mostrar el mapa
mapa
```
